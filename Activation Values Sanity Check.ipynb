```python
def tanh(z):
    g = ((np.exp(z)-np.exp(-z)) / (np.exp(z)+np.exp(-z)))
    return g
```

```python
activation_1 = tanh(10)
print(activation_1)
```

```python
def tanh(x):
    return np.tanh(x)

def tanh_deriv(x):
    return 1.0 - np.tanh(x)**2
```

```python
tanh_vals = []
tanh_deriv_vals = []

for i in range(-10,10,1):
    tanh_values = tanh(i)
    tanh_deriv_values = tanh_deriv(i)
    
    tanh_vals.append(tanh_values)
    tanh_deriv_vals.append(tanh_deriv_values)
    
    tanh_vals_df = pd.DataFrame(tanh_vals)
    tanh_deriv_vals_df = pd.DataFrame(tanh_deriv_vals)
```

```python
tanh_vals_df.plot()
```
